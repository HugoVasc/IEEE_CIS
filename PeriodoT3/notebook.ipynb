#### 2.1 Divisão entre treino e teste

```python
#split train test with numpy

import numpy as np


def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) #permutate the data
    test_set_size = int(len(data) * test_ratio) #calculate the test set size
    test_indices = shuffled_indices[:test_set_size] #get the test indices
    train_indices = shuffled_indices[test_set_size:] #get the train indices
    return data.iloc[train_indices], data.iloc[test_indices] #return the train and test sets
```

```python
X_train, X_test = split_train_test(x, 0.25)
```

```python
print("Train: ", X_train.shape, "Teste: ", X_test.shape)
```

```python
Y_train, Y_test = split_train_test(y, 0.25)
```

```python
print("Train: ", Y_train.shape, "Teste: ", Y_test.shape)
```

### 3ª Etapa: Definição do Modelo
- Função de ativação: ReLU (A função Relu foi escolhida pois apesar de ser não linear, está é diferenciàvel por partes, o que facilita o cálculo do gradiente.)


```python
class perceptron(object):
    def __init__(self, learning_rate=0.01, max_epochs=10):
        self.learning_rate = learning_rate
        self.max_epochs = max_epochs
        self.weights = np.random.random(30) #Gera um array de 30 posições com valores aleatórios entre 0 e 1
        self.bias = np.random.random(1) #Gera um valor aleatório entre 0 e 1
        self.history = []
    def __relu(self, x):
        return np.maximum(0,x) #Retorna 0 se x <= 0 e x se x > 0
    
    def __relu_derivative(self, x): #Retorna 0 se x <= 0 e 1 se x > 0
        x[x<=0] = 0
        x[x>0] = 1
        return x

    def cross_entropy(self, y, y_pred):
        return -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

    def cost(self, y, y_pred):
        return (y - y_pred)**2 #Retorna 1 se y != y_pred e 0 se y == y_pred
    
    def update_weights(self, x, y, y_pred):
        print("y_pred: ", y_pred.shape, "y: ", y.shape, "x: ", x.shape)
        self.weights += self.learning_rate * x * (y - y_pred)
        self.bias += self.learning_rate * (y - y_pred)

    def predict_one(self, x):
        output =  self.__relu(np.dot(x, self.weights) + self.bias)
        return 1 if output > 0 else 0
    
    def predict(self, x):
        y_pred = np.zeros(len(x))
        for i in range(len(x)):
            y_pred[i] = self.predict_one(x.iloc[i].to_numpy())
        return y_pred

    def fit(self, x, y):
        
        for epoch in range(self.max_epochs):
            y_pred = self.predict(x)
            self.update_weights(x, y, y_pred)
            loss = self.cost(y, y_pred)
            self.history.append(loss)
        return self
    
    
```

```python
model = perceptron(learning_rate=0.1, max_epochs=10)

model.fit(X_train, Y_train)

model.predict(X_test)
```
